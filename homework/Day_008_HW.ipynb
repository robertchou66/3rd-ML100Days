```python
import matplotlib.pyplot as plt
%matplotlib inline
age_ax = app_train['AMT_ANNUITY'].plot(kind='hist')
plt.show()
```

```python
mean1=np.mean(app_train['AMT_ANNUITY'])
std1 = np.std(app_train['AMT_ANNUITY'])
max1 = app_train['AMT_ANNUITY'].max()
ax = app_train['AMT_ANNUITY'].plot(kind='hist')
# plt.plot([0:250000],mean1, 'r-', label = 'prediction')
ax.hlines(y=mean1, xmin=0, xmax=max1, color='r',label = 'mean')
ax.hlines(y=std1, xmin=0, xmax=max1, color='g',label = 'std')
# plt.title("Assume we have data points (And the prediction)")
plt.legend(loc = 2)
plt.show()
```
